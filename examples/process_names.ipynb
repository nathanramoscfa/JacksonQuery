```python
pd.set_option('display.max_rows', None)
df = pd.concat([
    pd.Series(sorted(list(classification_schema.index))),
    pd.Series(sorted(list(management_fees.index)))
], keys=['classification_schema', 'management_fees'], axis=1)
```
